```python
# Importar las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo CSV
file_path = '/content/Precipitaciones_Totales_Mensuales.csv'
data = pd.read_csv(file_path)

# Verificar la estructura del dataframe
print(data.head())

# Crear un dataframe con las precipitaciones mensuales sumadas por cada municipio
monthly_precipitation = data.groupby('MUNICIPIO').sum(numeric_only=True).iloc[:, 7:] # Columnas de meses

# Encontrar el mes con mayor y menor precipitación total
monthly_totals = monthly_precipitation.sum()
max_precipitation_month = monthly_totals.idxmax()
min_precipitation_month = monthly_totals.idxmin()
max_precipitation_value = monthly_totals.max()
min_precipitation_value = monthly_totals.min()

# Municipios con la mayor y menor precipitación en esos meses
max_precipitation_city = monthly_precipitation[max_precipitation_month].idxmax()
min_precipitation_city = monthly_precipitation[min_precipitation_month].idxmin()

print(f"Mes con mayor precipitación: {max_precipitation_month} ({max_precipitation_value} mm)")
print(f"Municipio con mayor precipitación en {max_precipitation_month}: {max_precipitation_city}")

print(f"Mes con menor precipitación: {min_precipitation_month} ({min_precipitation_value} mm)")
print(f"Municipio con menor precipitación en {min_precipitation_month}: {min_precipitation_city}")

# Gráficas de precipitaciones por municipio (mayores y menores)
top_n = 10  # Número de municipios
municipios_sorted = monthly_precipitation.sum(axis=1).sort_values(ascending=False)

plt.figure(figsize=(14, 6))

# Municipios con mayores precipitaciones
plt.subplot(1, 2, 1)
municipios_sorted.head(top_n).plot(kind='bar', color='blue')
plt.title('Municipios con Mayores Precipitaciones')
plt.xlabel('Municipio')
plt.ylabel('Precipitaciones Totales (mm)')
plt.xticks(rotation=90)

# Municipios con menores precipitaciones
plt.subplot(1, 2, 2)
municipios_sorted.tail(top_n).plot(kind='bar', color='red')
plt.title('Municipios con Menores Precipitaciones')
plt.xlabel('Municipio')
plt.ylabel('Precipitaciones Totales (mm)')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# Gráficas de precipitaciones por mes (mayores y menores)
plt.figure(figsize=(14, 6))

# Meses con mayores precipitaciones
plt.subplot(1, 2, 1)
monthly_totals.sort_values(ascending=False).head(top_n).plot(kind='bar', color='green')
plt.title('Meses con Mayores Precipitaciones')
plt.xlabel('Mes')
plt.ylabel('Precipitaciones Totales (mm)')
plt.xticks(rotation=45)

# Meses con menores precipitaciones
plt.subplot(1, 2, 2)
monthly_totals.sort_values(ascending=True).head(top_n).plot(kind='bar', color='orange')
plt.title('Meses con Menores Precipitaciones')
plt.xlabel('Mes')
plt.ylabel('Precipitaciones Totales (mm)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Implementar el algoritmo de búsqueda DFS
def dfs(graph, start):
    stack = [start]
    visited = set()
    path = []

    while stack:
        current = stack.pop()
        if current not in visited:
            visited.add(current)
            path.append(current)
            for neighbor in graph[current]:
                stack.append(neighbor)

    return path

# Definir el grafo
graph = {
    'CiudadA': ['CiudadB', 'CiudadC'],
    'CiudadB': ['CiudadA', 'CiudadD'],
    'CiudadC': ['CiudadA', 'CiudadD'],
    'CiudadD': ['CiudadB', 'CiudadC']
}

# Encontrar el camino usando DFS
start_node = 'CiudadA'
dfs_path = dfs(graph, start_node)

print(f"Camino DFS desde {start_node}: {dfs_path}")
```
